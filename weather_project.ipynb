## 1.3. Thực hiện thu thập dữ liệu

Thêm các thư viện cần thiết cho việc thu thập

```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup as soup
from time import sleep
from datetime import datetime, timedelta
```

Xây dựng hàm thu thập dữ liệu từ website

```python
def get_weather_data(date):
    url = f"https://freemeteo.bg/weather/hanoi/history/daily-history/?gid=1581130&station=11376&date={date}&language=english&country=vietnam"
    res = requests.get(url)
    soup1 = soup(res.text, "html.parser")

    print(url)

    columns = ['time', 't1', 't2', 'wind_dir', 'wind_gust', 'hum', 'dew', 'pressure', 'icon', 'other', 'date']
    df = pd.DataFrame(columns=columns)

    for row in soup1.findAll("tr"):
        cells = row.findAll("td")
        if len(cells) == 10:
            time = cells[0].text.strip()
            t1 = cells[1].text.strip()
            t2 = cells[2].text.strip()
            wind_dir = cells[3].text.strip()
            wind_gust = cells[4].text.strip()
            hum = cells[5].text.strip()
            dew = cells[6].text.strip()
            pressure = cells[7].text.strip()
            icon = cells[8].span['data-icon'] if cells[8].span else ''
            other = cells[9].find('span', class_="details").text.strip() if cells[9].find('span', class_="details") else ''

            temp_df = pd.DataFrame([[time, t1, t2, wind_dir, wind_gust, hum, dew, pressure, icon, other, date]], columns=columns)
            df = pd.concat([df, temp_df], ignore_index=True)

    return df
```

Tạo một danh sách từ năm 01-01-2015 đến 31-05-2024 để lấy thông tin trong khoảng thời gian này. Để ứng với thời gian của dữ liệu 'aqi.csv'

```python
start_date = datetime.strptime("2015-01-01", "%Y-%m-%d")
end_date = datetime.strptime("2024-05-31", "%Y-%m-%d")
date_list = [(start_date + timedelta(days=x)).strftime("%Y-%m-%d") for x in range((end_date - start_date).days + 1)]
```

Lưu dữ liệu lại sau khi thu thập

```python
all_weather_data = pd.DataFrame()

for date in date_list:
    try:
        daily_weather_data = get_weather_data(date)
        all_weather_data = pd.concat([all_weather_data, daily_weather_data], ignore_index=True)
        print(f"Data for {date} added successfully.")
    except Exception as e:
        print(f"Failed to retrieve data for {date}: {str(e)}")

all_weather_data.to_csv("weather.csv", index=False, encoding='utf-8')
```

Tiến hành hợp nhất 2 nguồn dữ liệu thành một

```python
url1 = "weather.csv"
url2 = "aqi.csv"

df1 = pd.read_csv(url1)
df2 = pd.read_csv(url2)
```

```python
# Chuyen thanh datetime trong pd de merge lai voi nhau
df1['datetime'] = pd.to_datetime(df1['date'] + ' ' + df1['time'])
df2['Date (LT)'] = pd.to_datetime(df2['Date (LT)'])

# xu ly dupliate trong df2
df2_deduplicated = df2.drop_duplicates(subset='Date (LT)', keep='first')

# tao danh sach ngay thang nam
all_dates = pd.date_range(start='2015-01-01', end='2024-05-31', freq='D')
all_times = pd.date_range(start='00:00', end='23:00', freq='h').time

# tao dataframe tu danh sach ngay thang nam
df3 = pd.MultiIndex.from_product([all_dates, all_times], names=['date', 'time']).to_frame(index=False)
df3['datetime'] = pd.to_datetime(df3['date'].astype(str) + ' ' + df3['time'].astype(str))

# tien hanh hop nhat cac dataframe
result = pd.merge(df3, df1, how='left', left_on=['datetime'], right_on='datetime')
final_result = pd.merge(result, df2_deduplicated, how='left', left_on='datetime', right_on='Date (LT)')

# luu lai ket qua
final_result.to_csv("fin.csv", index=False, encoding='utf-8')
```

Xây dựng các hàm để sử dụng Gemini trong việc trích xuất feature

```python
def getT1(t1):
  response = model.generate_content(f"Ví dụ 't1=15°C' trả về '15'. Hãy trả về kết quả cho 't1={t1}'. Chỉ trả về theo chính xác mẫu tôi cung cấp")
  num = response.text.replace('•', '  *')
  return float(num)

def getWind(wind):
  response = model.generate_content(f"Dựa trên các ví dụ 'wind_dir=270°6 Km/h' trả về '270, 6', 'wind_dir=Variable at 2 Km/h' trả về '-1, 2', 'wind_dir=Calm' trả về (-1, 0) . Hãy trả về kết quả cho 'wind_dir={wind}'. Chỉ trả về theo chính xác mẫu tôi cung cấp")
  num = response.text.replace('•', '  *').split(',')
  return [float(num[0]), float(num[1])]

def getHum(hum):
  response = model.generate_content(f"Ví dụ 'hum=88%' trả về '88'. Hãy trả về kết quả cho 'hum={hum}'. Chỉ trả về theo chính xác mẫu tôi cung cấp")
  num = response.text.replace('•', '  *')
  return float(num)

def getDew(dew):
  response = model.generate_content(f"Ví dụ 'dew=13°C' trả về '13'. Hãy trả về kết quả cho 'dew={dew}'. Chỉ trả về theo chính xác mẫu tôi cung cấp")
  num = response.text.replace('•', '  *')
  return float(num)

def getPressure(pressure):
  response = model.generate_content(f"Ví dụ 'pressure=1022.0mb' trả về '1022.0'. Hãy trả về kết quả cho 'pressure={pressure}'. Chỉ trả về theo chính xác mẫu tôi cung cấp")
  num = response.text.replace('•', '  *')
  return float(num)
```

Xử lý các giá trị theo phương pháp xử lý string

---

Thay thế °C, %, mb thành rỗng và chuyển thành số thực, đồng thời thay thế -999 trong AQI = nan. Bởi vì -999 tương tự với việc không có dữ liệu AQI tại điểm dữ liệu đó

```python
df['t1'] = df['t1'].str.replace('°C', '').astype(float)
df['hum'] = df['hum'].str.replace('%', '').astype(float)
df['dew'] = df['dew'].str.replace('°C', '').astype(float)
df['pressure'] = df['pressure'].str.replace('mb', '').astype(float)
df['AQI'] = df['AQI'].replace(-999, np.nan)
```

Xử lý feature wind_dir, bởi vì feature này gồm hướng gió và tốc độ nên phải xử lý riêng và thêm một số trường hợp như đứng đó hoặc gió thay đổi hướng liên tục

```python
def parse_wind(data):
    if not isinstance(data, str):
        return (np.nan, np.nan)

    if data == "Calm":
        return (-1, 0)
    if "Variable" in data:
        speed = int(re.search(r'\d+', data).group())
        return (-1, speed)
    data = data.split(',')[0]
    match = re.match(r'(\d+)°(\d+)', data)
    if match:
        direction = int(match.group(1))
        speed = int(match.group(2))
        return (direction, speed)
    return (-1, -1)

df[['wind_direction', 'wind_speed']] = df['wind_dir'].apply(lambda x: pd.Series(parse_wind(x)))
```

Xử lý missing data bằng phương pháp KNN

## 3.1. Linear Regression

**Định nghĩa:**

Linear Regression (Hồi quy tuyến tính) là một kỹ thuật thống kê dùng để mô hình hóa và phân tích mối quan hệ giữa một biến phụ thuộc và một hoặc nhiều biến độc lập. Linear Regression chủ yếu được sử dụng để dự đoán giá trị của biến phụ thuộc dựa trên các giá trị của biến độc lập.

**Nguyên lý hoạt động:**

Linear Regression cố gắng tìm ra đường thẳng (hoặc mặt phẳng trong trường hợp nhiều biến độc lập) sao cho tổng bình phương các sai số giữa giá trị dự đoán và giá trị thực tế là nhỏ nhất. Đường thẳng này được gọi là đường hồi quy.

![image.png](https://realpython.com/cdn-cgi/image/width=950,format=auto/https://files.realpython.com/media/fig-lin-reg.a506035b654a.png)

**Công thức của Linear Regression đơn giản:**

\begin{align}
𝑌 = 𝛽_{0} + 𝛽_{1}𝑋 + 𝜖
\end{align}

Trong đó:
- 𝑌: Biến phụ thuộc (dependent variable).
- 𝑋: Biến độc lập (independent variable).
- $𝛽_{0}$: Hệ số chặn (intercept).
- $𝛽_{1}$: Hệ số góc (slope).
- 𝜖: Sai số (error term).


**Mục tiêu:**

Mục tiêu của Linear Regression là tìm ra đường hồi quy tốt nhất để dự đoán giá trị của biến phụ thuộc từ các biến độc lập. Điều này được thực hiện bằng cách tối ưu hóa các hệ số 𝛽 sao cho sai số dự đoán là nhỏ nhất.

![image.png](https://editor.analyticsvidhya.com/uploads/375512.jpg)

**Định nghĩa các tham số của model:**
- Intercept ($𝛽_{0}$): Giá trị dự đoán của biến phụ thuộc khi tất cả các biến độc lập bằng 0.
- Slope ($𝛽_{1}, 𝛽_{2},…,𝛽_{𝑛}$): Mức độ thay đổi của biến phụ thuộc khi biến độc lập thay đổi một đơn vị.
- Error term (𝜖): Phần không được giải thích bởi mô hình, bao gồm các yếu tố ngẫu nhiên hoặc không quan sát được.

## 3.2. Support Vector Regression (SVR)

**Định nghĩa:**
- **Support Vector Regression (SVR)** là một loại của thuật toán Support Vector Machine (SVM) và thường được dùng để phân tích hồi quy.
- SVR sử dụng khái niệm siêu phẳng (hyperplane) và lề (margin).
- Trong SVR, **lề** được định nghĩa là khả năng chịu lỗi của mô hình, còn được gọi là **ε-insensitive tube** (ống không nhạy cảm ε):
    + Ống này cho phép một số điểm dữ liệu sai lệch so với siêu phẳng mà không bị tính là lỗi.
    + Các đường lề được chọn sao cho chúng bao phủ tất cả dữ liệu (hard margin) hoặc cho phép một số vi phạm (soft margin).
- **Siêu phẳng** phù hợp nhất có thể với dữ liệu nằm trong ε-insensitive tube.

![image.png](https://i.imgur.com/HaPW8d2.png)

- **Mục tiêu của SVR** là tìm ra sự phù hợp nhất để dự đoán chính xác biến mục tiêu đồng thời giảm độ phức tạp để tránh bị overfitting.
- Trong trường hợp tối ưu hóa không khả thi, ta cho phép xảy ra một số lỗi => sinh ra các **biến chùng (slack variable)** - những điểm nằm ngoài ε-insensitive tube.
- Để cân bằng sự cân bằng giữa độ phức tạp của mô hình (tức là độ phẳng của f(x)) và tổng độ lệch ngoài ε-insensitive tube => sử dụng **tham số chính quy hóa 𝐶** > 0.
- SVR ấn định sai số dự đoán bằng 0 cho các điểm nằm bên trong ε-insensitive tube, trong khi nó xử phạt các biến chùng tỷ lệ với ξ của chúng (Khoảng cách của các biến chùng từ ranh giới của ε-insensitive tube được biểu thị bằng ξ).

![image.png](https://i.imgur.com/FWE9Mi7.png)

**Định nghĩa các tham số của model** là kernel function, C và ε:
- **kernel function** được sử dụng ở đây là 'rbf': hàm Gaussian (radial basis function). Kernel function xác định cách mà không gian đầu vào được ánh xạ vào không gian đặc trưng cao hơn, cho phép mô hình xử lý các mối quan hệ phi tuyến tính giữa các điểm dữ liệu.
- **C**: hằng số dương giúp điều chỉnh mức độ cho phép sai số trên tập huấn luyện, cân đối giữa độ lớn của margin và sự hy sinh của các điểm nằm ngoài ε-insensitive tube.
- **ε**: xác định khả năng chịu lỗi của mô hình bằng cách xác định một khoảng lỗi (ε-insensitive tube) mà trong đó không có phạt lỗi cho các điểm dữ liệu nằm trong khoảng này.

## 3.3. Random Forest (RF)

**Định nghĩa:**

Random Forest là một thuật toán học máy thuộc nhóm ensemble learning, được sử dụng cho cả bài toán phân loại và hồi quy. Random Forest tạo ra một "rừng" gồm nhiều cây quyết định (Decision trees) và kết hợp các kết quả của chúng để cải thiện độ chính xác và kiểm soát overfitting.

**Nguyên lý hoạt động:**

Random Forest hoạt động bằng cách xây dựng nhiều cây quyết định trong quá trình huấn luyện và dự đoán đầu ra dựa trên kết quả trung bình (đối với hồi quy) hoặc kết quả đa số (đối với phân loại) từ tất cả các cây.

![image.png](https://anasbrital98.github.io/assets/img/20/random-forest.jpg)

**Công thức tổng quát của mô hình Random Forest:**

\begin{align}
\hat{Y} = \frac{1}{N} \sum_{i=1}^{n} T_{i}(x)
\end{align}

Trong đó:
- $\hat{Y}$: Giá trị dự đoán cuối cùng
- $N$: Số lượng cây trong rừng
- $T_{i}(x)$: Kết quả dự đoán từ cây thứ i

**Mục tiêu:**


Mục tiêu của Random Forest là tăng cường độ chính xác của dự đoán và giảm thiểu overfitting bằng cách sử dụng nhiều cây quyết định. Mỗi cây được xây dựng từ một mẫu ngẫu nhiên của dữ liệu huấn luyện và sử dụng một tập hợp ngẫu nhiên các tính năng để tách tại mỗi nút.

**Các tham số của mô hình:**
- Number of trees (n_estimators): Số lượng cây trong rừng. Số lượng cây
lớn thường dẫn đến độ chính xác cao hơn nhưng cũng tăng thời gian tính toán.
- Maximum depth (max_depth): Độ sâu tối đa của mỗi cây. Độ sâu lớn có thể dẫn đến overfitting, trong khi độ sâu nhỏ có thể dẫn đến underfitting.
- Minimum samples split (min_samples_split): Số lượng mẫu tối thiểu cần thiết để tách một nút.
- Minimum samples leaf (min_samples_leaf): Số lượng mẫu tối thiểu cần thiết để tạo thành một lá.
- Bootstrap: Quyết định xem có sử dụng bootstrap samples (mẫu ngẫu nhiên với thay thế) khi xây dựng cây hay không.

## 3.4. Multi-layer Perceptron (MLP)

**Định nghĩa:**

Multi-layer Perceptron (MLP) là một loại mạng nơ-ron nhân tạo (artificial neural network) được sử dụng rộng rãi trong học sâu (deep learning). MLP bao gồm nhiều lớp nơ-ron (neurons) có thể học và mô hình hóa các mối quan hệ phức tạp giữa các biến đầu vào và đầu ra.

**Nguyên lý hoạt động:**

MLP hoạt động dựa trên cấu trúc gồm nhiều lớp nơ-ron: một lớp đầu vào, một hoặc nhiều lớp ẩn, và một lớp đầu ra. Mỗi nơ-ron trong một lớp được kết nối với mỗi nơ-ron trong lớp tiếp theo. Mỗi kết nối này có một trọng số (weight) và mỗi nơ-ron có một ngưỡng (bias).

![image.png](https://miro.medium.com/v2/resize:fit:650/1*xPbk6HNUgXSWS0sRSTXLwg.png)

**Công thức tổng quát của một nơ-ron:**

\begin{align}
y = f(\sum_{i=1}^{n}w_{i}x_{i}+b)
\end{align}

Trong đó:
- $y$: Đầu ra của nơ-ron
- $f$: Hàm kích hoạt (activation function), thường là hàm phi tuyến
- $w_{i}$: Trọng số của kết nối từ nơ-ron thứ 𝑖 của lớp trước
- $x_{i}$: Đầu vào từ nơ-ron thứ 𝑖 của lớp trước
- $b$: Ngưỡng của nơ-ron

**Mục tiêu:**

Mục tiêu của MLP là học các trọng số và ngưỡng sao cho mạng có thể dự đoán chính xác đầu ra từ các đầu vào thông qua quá trình huấn luyện, sử dụng phương pháp lan truyền ngược (backpropagation) để tối thiểu hóa hàm mất mát (loss function).

**Các tham số của mô hình:**
- Number of layers (n_layers): Số lượng lớp trong mạng. Một mạng đơn giản có ít nhất ba lớp: đầu vào, một lớp ẩn và đầu ra. Các mạng phức tạp hơn có thể có nhiều lớp ẩn.
- Number of neurons per layer (n_neurons): Số lượng nơ-ron trong mỗi lớp ẩn.
- Activation function: Hàm kích hoạt được sử dụng trong các nơ-ron, phổ biến nhất là ReLU (Rectified Linear Unit), sigmoid và tanh.
- Learning rate: Tốc độ học, điều chỉnh mức độ cập nhật trọng số trong quá trình huấn luyện.
- Epochs: Số lần toàn bộ tập dữ liệu huấn luyện được sử dụng để cập nhật trọng số của mạng.
- Batch size: Số lượng mẫu dữ liệu được sử dụng để cập nhật trọng số trong mỗi lần cập nhật.

## 3.5. Các thông số đánh giá

$\text{MSE}$ (Mean Squared Error):
- $\text{MSE}$ là số liệu phổ biến nhất được sử dụng cho các bài toán hồi quy.
- $\text{MSE}$ thể hiện sai số bình phương trung bình giữa các giá trị được dự đoán và thực tế.
- $\text{MSE}$ cho biết mức độ gần của đường hồi quy với một tập hợp các điểm.
- Giảm $\text{MSE}$ thì mô hình có thể chính xác hơn, có nghĩa là mô hình gần với dữ liệu thực tế hơn.
$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^n (Y_i - \hat{Y}_i)^2
$$
trong đó n là số điểm dữ liệu, $Y_i$ là giá trị quan sát và $\hat{Y}_i$ là giá trị dự đoán


$R^2$ (R-squared):
- $R^2$ hay còn gọi là hệ số xác định đo lường độ biến thiên của biến phụ thuộc $Y$ đang được giải thích bởi các biến độc lập $X_i$ trong mô hình hồi quy.
- $R^2$ càng lớn thì giá trị tổng bình phương sai số càng nhỏ, dự báo càng tốt.
$$
R^2 = 1 - \frac{SSR}{SST}
$$
trong đó 
$$
\text{SSR} = \sum_{i=1}^n (Y_i - \hat{Y}_i)^2
$$
và 
$$
\text{SST} = \sum_{i=1}^n (Y_i - \bar{Y})^2
$$

### 4.2.4. Multi-layer Perceptron

Mô hình MLP được xem như một thử nghiệm nhỏ. Xem xét việc ứng dụng neural network vào các tác vụ, cụ thể trong trường hợp này là hồi quy. Nên chỉ thử nghiệm MLP thử nghiệm trên kịch bản số 6

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import pandas as pd
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
from sklearn.metrics import r2_score
```

Chuyển dữ liệu thành kiểu tensor để có thể sử dụng trong việc train model bằng pytorch

```python
X_torch = torch.tensor(X.values, dtype=torch.float32)
y_torch = torch.tensor(y.values, dtype=torch.float32)

dataset = TensorDataset(X_torch, y_torch)
```

Chia tập dữ liệu thành 3 phần: train, validate, test tương ứng 50%, 20%, 30% với dữ liệu gốc. Chọn size test là 30% để dễ dàng so sánh các mô hình đã triển khai.

```python
total_size = len(dataset)
train_size = int(0.5 * total_size)
val_size = int(0.2 * total_size)

train_X = X_torch[:train_size]
train_y = y_torch[:train_size]
val_X = X_torch[train_size:train_size + val_size]
val_y = y_torch[train_size:train_size + val_size]
test_X = X_torch[train_size + val_size:]
test_y = y_torch[train_size + val_size:]
```

Tạo data loader

```python
train_dataset = TensorDataset(train_X, train_y)
test_dataset = TensorDataset(test_X, test_y)
val_dataset = TensorDataset(val_X, val_y)

batch = 4
train_loader = DataLoader(train_dataset, batch_size=batch, shuffle=False)  # Không xáo trộn dữ liệu chuỗi thời gian
test_loader = DataLoader(test_dataset, batch_size=batch, shuffle=False)
val_loader = DataLoader(val_dataset, batch_size=batch, shuffle=False)
```

```python
# Sử dụng gpu để train model (nếu có)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
device
```

Khởi tạo mô hình MLP với các tham số như sau:


*   Số hidden layer: 5
*   Số unit của mỗi hidden layer: 32, 64, 128, 256, 128
*   Hàm kích hoạt: ReLU cho mỗi hiden layer
*   Hàm tối ưu sử dụng: Adam
*   Hàm loss: MSE
*   Tốc độc học: learning_rate = 0.001
*   Batch size: 4
*   Epoch: 100




```python
class MLP(nn.Module):
    def __init__(self):
        super(MLP, self).__init__()
        self.layer1 = nn.Linear(6, 32)
        self.layer2 = nn.Linear(32, 64)
        self.layer3 = nn.Linear(64, 128)
        self.layer4 = nn.Linear(128, 256)
        self.layer5 = nn.Linear(256, 128)
        self.layer6 = nn.Linear(128, 1)


    def forward(self, x):
        x = F.relu(self.layer1(x))
        x = F.relu(self.layer2(x))
        x = F.relu(self.layer3(x))
        x = F.relu(self.layer4(x))
        x = F.relu(self.layer5(x))
        return self.layer6(x)

model = MLP()
criterion = nn.MSELoss(reduction='mean')
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
val_MSE = []
model.to(device)

for epoch in range(100):
    model.train()
    for data, label in train_loader:
        data, label = data.to(device), label.to(device)
        output = model(data)
        loss = criterion(output, label)
        # zero the parameter gradients
        optimizer.zero_grad()
        # forward + backward + optimize
        loss.backward()
        optimizer.step()
    if(epoch % 10 == 0):
      print(f'Progress {epoch} %')

    model.eval()
    val_loss = 0.
    with torch.no_grad():
        for data, label in val_loader:
            data, label = data.to(device), label.to(device)
            output = model(data)
            loss = criterion(output, label)
            val_loss += loss.item()
        val_MSE.append(val_loss / len(val_loader))
    model.train()

    if len(val_MSE) == 0 or val_MSE[-1] <= min(np.array(val_MSE)):
        print("Best model on epoch: {}, val_mse: {:.4f}".format(epoch, val_MSE[-1]))
        torch.save(model.state_dict(), "Regression-best.th")
```

```python
modelb = MLP().to(device)
modelb.load_state_dict(torch.load('Regression-best.th', map_location=device))

all_targets = []
all_predictions = []

modelb.eval()
test_loss, test_step = 0, 0

with torch.no_grad():
    for data, label in test_loader:
        data, label = data.to(device), label.to(device)
        output = modelb(data)
        loss = criterion(output, label)
        test_loss += loss.item()
        all_targets.extend(label.cpu().numpy())
        all_predictions.extend(output.cpu().numpy())

# tính toán mse và visualize loss theo epoch
print(f"Mse of the best model on the test data is: {test_loss / len(test_loader)}")

plt.figure(figsize=(10, 5))
plt.plot(val_MSE, label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss Curves')
plt.legend()
plt.show()

# Tính toán r2
r2 = r2_score(all_targets, all_predictions)
print(f'R2: {r2:.4f}')
```

* Nhận thấy tầm quan trọng của dự đoán AQI trong việc bảo vệ sức khỏe cộng đồng. Hiểu rõ hơn về tác động thực tiễn của công việc mình làm.
* Cải thiện khả năng phân tích dữ liệu và ra quyết định dựa trên kết quả mô hình.
* Hoạt động và làm việc nhóm đạt hiệu quả cao

### Các khó khăn

Tuy nhiên trong quá trình làm việc, nhóm cũng gặp không ít khó khăn như:
* Gặp vấn đề trong việc tìm kiếm các trang web có bộ dữ liệu phù hợp.
* Xử lý các giá trị thiếu và dữ liệu không đồng nhất, đòi hỏi nhiều công sức và thời gian.
* Cần phải làm sạch và chuẩn hóa dữ liệu một cách cẩn thận để đảm bảo độ chính xác của mô hình.
* Khó khăn trong việc lựa chọn mô hình phù hợp cho dữ liệu.
* Quá trình tối ưu hóa tham số và huấn luyện mô hình phức tạp và tốn nhiều thời gian.
* Cần thực hiện nhiều thử nghiệm để đảm bảo kết quả đánh giá đáng tin cậy.
